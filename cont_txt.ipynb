##1. Monte uma lista de palavras Ãºnicas ordenadas, removendo caracteres de pontuaÃ§Ã£o, espaÃ§os em branco e quebras de linha


```
with open("/content/1984.txt") as arquivo:  # abrir o arquivo e automatizar o close
  leitura = arquivo.read()  # ler o arquivo
  pontuacao=("\n","!","'","(",")",",","-",".",":",";","?") # caracteres de pontuaÃ§Ã£o que serÃ£o removidos
  textolimpo = leitura

  for i in range(len(pontuacao)):  # para cada item i de pontuaÃ§Ã£o
    textolimpo=textolimpo.replace(pontuacao[i],'')  # remover o item i e substituir por vazio no texto

textolimpo=textolimpo.lower() #  todas as letras minÃºsculas
textolimpo=textolimpo.strip() #  remove espaÃ§os em branco
novalista=textolimpo.split() #  criar uma lista com as palavras
novalista=sorted(novalista) #  ordenando a lista de palavras
#print(novalista)  # aqui a lista possui palavras repetidas

from collections import Counter  # contador de elementos
n=Counter(novalista)  #  quantas vezes cada palavra aparece
#print('quantas vezes cada palavra aparece: \n',n) 

palavras=list(n)  # criar uma lista com cada palavra diferente
print('\nLista de palavras Ãºnicas ordenadas: ',palavras,'\nQuantidade de palavras diferentes: ',len(n)) 
```

##4. Calcule a frequÃªncia de cada caracter (percentual).

```
m=Counter(leitura)  #  quantas vezes cada caracter aparece
caracter=list(m.items())  # criar uma lista com cada caracter
#print(caracter)

print("frequÃªncia de cada caracter (percentual) Ã©: ")
for i in range(len(caracter)):  # para cada caracter
  print(list(caracter[i][0]),"=",round(caracter[i][1]/(len(leitura))*100,4))  # relacionar o primeiro com o segundo respectivo elemento em percentual com 4 casas decimais
  i+=1
```
